# BSR 4s step-wise excitation

```python
filename = "./crs/BSR.1s.step-excite.txt"
bsr = cross.multipleCrossSections(filename)

label4s = ["Ar(4s'[1/2]0)", "Ar(4s'[1/2]1)", "Ar(4s[3/2]1)", "Ar(4s[3/2]2)"]

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
    "font.size": 20,
})

for k, label in enumerate(label4s):
    plt.figure(k)
    for c in bsr.crs:
        tmp = c.colName.split(" ")
        reactant, product = tmp[0], tmp[-1]
        if ((reactant == label) and (product not in label4s)):
            plt.loglog(c.data[:,0], c.data[:,1])
    #     print("type: %s" % cross.typeDictI2S[c.colType])
#         print("name: %s" % c.colName)
    #     print(c.colName.split(" "))
    plt.title("%s to 4p levels" % label)
    plt.xlabel("$\epsilon$ ($eV$)")
    plt.ylabel("$\sigma$ ($m^2$)")
    fig = plt.gcf()
    fig.set_size_inches(9, 7)
```
