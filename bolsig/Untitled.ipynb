```python
from scipy.interpolate import interp1d
plt.rcParams['font.size'] = '16'

# E0 = 11.548
E0 = 11.62
x = np.linspace(np.log(5.0e-3),np.log(80),201)
x = np.exp(x) + E0
print(min(x),max(x))

mean, var = np.zeros(len(x)), np.zeros(len(x))

plt.figure(1)
h = []
for k, dataset in enumerate(datasets[:-2]):
    if ((k==4) or (k==5)):
#     if (not (k==3)):
        continue
# dataset = datasets[7]
    filename = "./crs/%s.txt" % dataset
    tmp = cross.multipleCrossSections(filename)

    for c in tmp.crs:
#         if ((c.colType==2) and (c.deltaE<11.6)):
        if ((c.colType==2) and (c.deltaE>11.6) and (c.deltaE<11.7)):
            plt.loglog(c.data[:,0]-c.deltaE,c.data[:,1],'-')
            print(c.data[0,0],max(c.data[:,0]-E0))
#             mask = ((np.log(c.data[:,0] - E0)>-10.0))
            f = interp1d(c.data[1:,0], c.data[1:,1], kind='linear')
            
            lowx = (x<c.data[1,0])
            fx = (x[lowx] - E0) / (c.data[1,0] - E0) * c.data[1,1]
            highx = (x>=c.data[1,0])
            fx = np.append(fx, f(x[highx]))
            
#             plt.loglog(x-E0,fx,'-')
            mean += fx
            var += fx ** 2
            h += [dataset]
mean /= 5.0
var /= 5.0
var -= mean ** 2
            
plt.xlim([1e-3, 1e3])
plt.title('$4s[3/2]^o_2$ ($11.62eV$)')
plt.xlabel('$E-E_0$ ($eV$)')
plt.ylabel('Cross section ($m^2$)')
fig = plt.gcf()
fig.set_size_inches(6, 5)
plt.tight_layout()
plt.legend(h,bbox_to_anchor=(.7, .5))
plt.savefig('./excitation2.lxcat.png',dpi=500)
plt.show()

plt.figure(2)
plt.loglog(x-E0,mean,'-')

fig3 = plt.figure(3)
plt.semilogx(x-E0,np.sqrt(var) / mean * 100,'-')
plt.ylim([0, 100])
plt.xlabel('$E-E_0$ ($eV$)')
plt.ylabel('Std. Dev. ($\%$)')
fig3.set_size_inches(6, 5)
plt.tight_layout()
plt.savefig('./excitation2.lxcat.std.png',dpi=500)
```

```python
from scipy.interpolate import interp1d
plt.rcParams['font.size'] = '16'

x = np.linspace(np.log(5.0e-3),np.log(150),201)
x = np.exp(x)
print(min(x),max(x))

mean, var = np.zeros(len(x)), np.zeros(len(x))

plt.figure(1)
h = []
for k, dataset in enumerate(datasets[:-2]):
    if ((k==4) or (k==5)):
        continue
# dataset = datasets[7]
    filename = "./crs/%s.txt" % dataset
    tmp = cross.multipleCrossSections(filename)

    for c in tmp.crs:
        if ((c.colType==0) or (c.colType==1)):
#             plt.loglog(c.data[:,0]-c.deltaE,c.data[:,1],'-')
            print(c.data[0,0],max(c.data[:,0]))
#             mask = ((np.log(c.data[:,0] - E0)>-10.0))
            f = interp1d(c.data[:,0], c.data[:,1], kind='linear')
            fx = f(x)
            plt.loglog(x,fx,'-')
            mean += fx
            var += fx ** 2
            h += [dataset]
mean /= 5.0
var /= 5.0
var -= mean ** 2
            
plt.title('Momentum transfer')
plt.xlabel('Electron energy ($eV$)')
plt.ylabel('Cross section ($m^2$)')
fig = plt.gcf()
fig.set_size_inches(6, 5)
plt.tight_layout()
plt.legend(h,bbox_to_anchor=(.7, .5))
plt.savefig('./momentum.lxcat.png',dpi=500)
plt.show()

plt.figure(2)
plt.loglog(x,mean,'-')

fig3 = plt.figure(3)
plt.semilogx(x,np.sqrt(var) / mean * 100,'-')
plt.ylim([0, 40])
plt.xlabel('$E$ ($eV$)')
plt.ylabel('Std. Dev. ($\%$)')
fig3.set_size_inches(6, 5)
plt.tight_layout()
plt.savefig('./momentum.lxcat.std.png',dpi=500)
```
