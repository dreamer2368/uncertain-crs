```python
import numpy as np
import matplotlib.pyplot as plt
from interp import quadratic_interpolate

%matplotlib inline
```

```python
Ng, L = 100, 2.0 * np.pi

xg = np.linspace(0.0, L, Ng) - 1.0
yg = np.sin(xg)
dx = L / (Ng - 1)

Np = 1000
xp = np.linspace(0.0, L, Np) - 1.0
yp = np.zeros(Np,)
grad_yp = np.zeros(Np,)
for k in range(Np):
    yp[k], grad_yp[k] = quadratic_interpolate(xg, yg, dx, xp[k])


plt.figure(1)
plt.plot(xg,yg,'-b')
plt.plot(xp,yp,'--r')
plt.plot(xp,grad_yp,'-.g')
```
