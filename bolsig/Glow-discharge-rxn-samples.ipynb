# 1s-lumped excitation

```python
rootDir = 'glow-discharge/rxn-samples.1'
rateName = '1s-lumped'
nSample = 7200
nPts = 200
rateSamples = np.zeros([nSample, nPts])

for n in range(nSample):
    filename = '%s/%s.%08d.h5' % (rootDir, rateName, n)
    with h5py.File(filename,'r') as f:
        rateSamples[n] = f['table'][:,1]
        Te = f['table'][:,0]

sig1 = int(0.5 * (1.0 - 0.68) * nSample)
sig2 = int(0.5 * (1.0 - 0.95) * nSample)
sig3 = int(0.5 * (1.0 - 0.997) * nSample)
median = int(0.5 * nSample)

rateSamples.sort(axis = 0)

plt.loglog(Te, rateSamples[median,:],'-', color='brown', linewidth=2.0)
plt.fill_between(Te, rateSamples[-sig3,:], rateSamples[sig3,:], color='brown', alpha=0.2)
plt.fill_between(Te, rateSamples[-sig2,:], rateSamples[sig2,:], color='brown', alpha=0.3)
plt.fill_between(Te, rateSamples[-sig1,:], rateSamples[sig1,:], color='brown', alpha=0.4)

filename = 'glow-discharge/%s.sample_intervals.txt' % rateName
data = np.array([Te, rateSamples[median,:], rateSamples[-sig1,:], rateSamples[sig1,:],
                                            rateSamples[-sig2,:], rateSamples[sig2,:],
                                            rateSamples[-sig3,:], rateSamples[sig3,:]]).T
np.savetxt(filename, data)
```

# Transport

```python
rootDir = 'glow-discharge/rxn-samples.1'
rateName = 'Transport'
nSample = 7200
nPts = 200
mobilitySamples = np.zeros([nSample, nPts])
diffusivitySamples = np.zeros([nSample, nPts])

for n in range(nSample):
    filename = '%s/%s.%08d.h5' % (rootDir, rateName, n)
    with h5py.File(filename,'r') as f:
        mobilitySamples[n] = f['mobility'][:,1]
        diffusivitySamples[n] = f['diffusivity'][:,1]
        Te = f['mobility'][:,0]

sig1 = int(0.5 * (1.0 - 0.68) * nSample)
sig2 = int(0.5 * (1.0 - 0.95) * nSample)
sig3 = int(0.5 * (1.0 - 0.997) * nSample)
median = int(0.5 * nSample)

mobilitySamples.sort(axis = 0)
diffusivitySamples.sort(axis = 0)

plt.figure(1)
plt.loglog(Te, mobilitySamples[median,:],'-', color='brown', linewidth=2.0)
plt.fill_between(Te, mobilitySamples[-sig3,:], mobilitySamples[sig3,:], color='brown', alpha=0.2)
plt.fill_between(Te, mobilitySamples[-sig2,:], mobilitySamples[sig2,:], color='brown', alpha=0.3)
plt.fill_between(Te, mobilitySamples[-sig1,:], mobilitySamples[sig1,:], color='brown', alpha=0.4)

plt.figure(2)
plt.loglog(Te, diffusivitySamples[median,:],'-', color='brown', linewidth=2.0)
plt.fill_between(Te, diffusivitySamples[-sig3,:], diffusivitySamples[sig3,:], color='brown', alpha=0.2)
plt.fill_between(Te, diffusivitySamples[-sig2,:], diffusivitySamples[sig2,:], color='brown', alpha=0.3)
plt.fill_between(Te, diffusivitySamples[-sig1,:], diffusivitySamples[sig1,:], color='brown', alpha=0.4)

filename = 'glow-discharge/%s.sample_intervals.txt' % 'mobility'
data = np.array([Te, mobilitySamples[median,:], mobilitySamples[-sig1,:], mobilitySamples[sig1,:],
                                            mobilitySamples[-sig2,:], mobilitySamples[sig2,:],
                                            mobilitySamples[-sig3,:], mobilitySamples[sig3,:]]).T
np.savetxt(filename, data)

filename = 'glow-discharge/%s.sample_intervals.txt' % 'diffusivity'
data = np.array([Te, diffusivitySamples[median,:], diffusivitySamples[-sig1,:], diffusivitySamples[sig1,:],
                                                    diffusivitySamples[-sig2,:], diffusivitySamples[sig2,:],
                                                    diffusivitySamples[-sig3,:], diffusivitySamples[sig3,:]]).T
np.savetxt(filename, data)
```
